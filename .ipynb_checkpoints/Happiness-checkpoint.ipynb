```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Users\tborn\OneDrive\Desktop\world_happiness\world-happiness-report.csv")
display(df.head(30))

print(df.isna().sum())

                
```

```python
for col in df.select_dtypes(include=['float64', 'int64']).columns:
    plt.figure(figsize=(8, 1.5))  # wide and compact for horizontal boxplot
    sns.boxplot(x=df[col])
    plt.title(f'Boxplot: {col}', fontsize=12)
    plt.xlabel(col)
    plt.tight_layout()
    plt.show()
```

```python
for col in df.select_dtypes(include=['float64', 'int64']):
    df[col] = df[col].fillna(df[col].median())
```
